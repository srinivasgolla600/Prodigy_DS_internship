```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
da= pd.read_csv("/content/drive/MyDrive/DATA_sets/world_population.csv")
```

```python
top_ten = da.sort_values(by="2022 Population",ascending =False).head(10)
```

```python
top_ten
```

```python
top_fig = px.bar(top_ten,x="Country",y="2022 Population",color="Rank",title="Top Ten Countries by Population in 2022",hover_data= top_ten.columns)

top_fig.show()
```

```python
bottom_ten = da.sort_values(by="2022 Population",ascending=False).tail(10)
```

```python
bottom_ten
```

```python
bottom_fig = px.bar(bottom_ten,x="Country",y="2022 Population",color="Growth Rate",title="Top Ten Countries by Population in 2022",hover_data= top_ten.columns)
bottom_fig.show()
```

```python
px.histogram(top_ten,y="2022 Population",x="Country",color="World Population Percentage")
```
